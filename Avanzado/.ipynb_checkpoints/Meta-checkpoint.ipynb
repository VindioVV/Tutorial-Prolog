```python
import ipywidgets as widgets
from PIL import Image
import io

error_images_ejemplo_1 = [
    '../Basico/imagenes/error/error_image_1.png',
    '../Basico/imagenes/error/error_image_2.png',
    '../Basico/imagenes/error/error_image_3.png',
    '../Basico/imagenes/error/error_image_4.png',
    '../Basico/imagenes/error/error_image_5.png'
]

def create_button(description, color, font_size, font_weight, width='auto'):
    button = widgets.Button(
        description=description,
        layout=widgets.Layout(width=width, padding='2px', margin='5px'),
        style={'button_color': color, 'font_weight': font_weight, 'font_size': font_size, 'text_align': 'center'}
    )
    return button


def display_image(image_path, format='png', width='300px', height='200px'):
    with open(image_path, "rb") as f:
        img = Image.open(f)
        img_byte_array = io.BytesIO()
        img.save(img_byte_array, format=format)
        img_data = img_byte_array.getvalue()
    
    return widgets.Image(value=img_data, format=format, layout=widgets.Layout(width=width, height=height))


class BaseQuizApp:
    def __init__(self, quiz_data):
        self.quiz_data = quiz_data
        self.current_question_index = 0
        self.error_count = 0
        self.output_box = widgets.Output()
        self.create_quiz()
        self.display()

    def create_quiz(self):
        pass

    def display(self):
        with self.output_box:
            self.output_box.clear_output()
            display(self.question_label)
            display(self.image_box)  
            display(self.options_box)

class QuizApp_ejemplo_1(BaseQuizApp):
    def __init__(self, quiz_data):
        super().__init__(quiz_data)

    def create_quiz(self):
        self.output_box.clear_output()
        self.option_buttons = []

        if self.current_question_index < len(self.quiz_data):
            quiz = self.quiz_data[self.current_question_index]

            self.question_label = widgets.HTML(
                value=f"<h2 style='color: #0056b3; text-align: center;'>{quiz['question']}</h2>"
            )

            self.option_buttons = [
                create_button(option, '#e8ebe9', '14px', 'bold') for option in quiz['options']
            ]

            for button in self.option_buttons:
                button.on_click(lambda b, opt=button.description: self.on_button_clicked(opt, quiz['correct_option']))

            with open('../Basico/imagenes/base/base.png', 'rb') as f:
                self.image_widget = widgets.Image(
                    value=f.read(),
                    format='png', 
                    layout=widgets.Layout(max_width='700px', max_height='700px', object_fit='contain')
                )
            self.image_box = widgets.HBox([self.image_widget], layout=widgets.Layout(justify_content='center'))

            self.options_box = widgets.VBox(self.option_buttons, layout=widgets.Layout(align_items='center'))

        else:
            self.question_label = widgets.HTML(
                value="<h2 style='color: #28a745; text-align: center;'>¡Bien hecho!</h2>"
            )
            self.options_box = widgets.VBox([])
            self.image_box = widgets.VBox([])

    def on_button_clicked(self, option, correct_option):
        if option == correct_option:
            self.show_message("Correcto", "¡Buena!")
            self.next_question()
        else:
            self.error_count += 1
            message = self.get_error_message()
            self.show_error_message(message)

    def get_error_message(self):
        if self.error_count < 2:
            return "Whops, esa no era la respuesta correcta. Prueba otra vez."
        elif self.error_count < 4:
            return "¿Estás seguro de que has atendido a la explicación?"
        elif self.error_count < 6:
            return "Estás haciendo al búho llorar."
        elif self.error_count < 7:
            return "¿Te diviertes?"
        else:
            return "..."

    def show_message(self, title, message):
        with self.output_box:
            self.output_box.clear_output()
            display(widgets.HTML(f"<h3 style='text-align: center;'>{title}: {message}</h3>"))

    def show_error_message(self, message):
        with self.output_box:
            self.output_box.clear_output()
    
            error_img_path = self.get_error_image_path()
    
            with open(error_img_path, "rb") as file:
                error_img = widgets.Image(
                    value=file.read(),
                    format='png',  
                    layout=widgets.Layout(max_width='500px', max_height='400px', object_fit='contain')
                )
    
            error_button = create_button("Venga voy", '#f44336', '14px', 'bold')
            error_button.on_click(self.retry_current_question)
    
            display(widgets.VBox([
                widgets.HBox([error_img], layout=widgets.Layout(justify_content='center')),
                widgets.HTML(f"<h4 style='text-align: center;'>{message}</h4>"),
                error_button
            ], layout=widgets.Layout(align_items='center')))

    def retry_current_question(self, b):
        self.create_quiz()
        self.display()

    def get_error_image_path(self):
        index = min(self.error_count - 1, len(error_images_ejemplo_1) - 1)
        return error_images_ejemplo_1[index]

    def next_question(self):
        self.current_question_index += 1
        self.create_quiz()
        self.display()

def run_quiz(quiz_data):
    quiz_app = QuizApp_ejemplo_1(quiz_data)
    display(quiz_app.output_box)
```

<h1 id="metapredicados-y-predicados-de-control">Metapredicados y predicados de control</h1>
<p>En Prolog tenemos una serie de herramientas muy poderosas y útiles ya definidas en la propia librería interna de Prolog, una serie de &quot;funciones&quot; intrínsecas del lenguaje como sucede con por ejemplo Java. Las podemos dividir entre <strong>metapredicados</strong>(No, no son accionistas de Zuckerberg, se llaman <strong>meta</strong> porque sus argumentos son otros predicados. Vamos, <a href="https://www.youtube.com/watch?v=SsShSms06-8">predicados dentro de otros predicados</a>), <strong>predicados de control</strong> para influenciar el flujo que seguirá el motor de inferencia(Ya has visto en capítulos anteriores el uso del <strong>&quot;No se puede demostrar&quot;</strong>) o los predicados sobre la base de hechos(Que nos permiten añadir o eliminar hechos). Hablaremos ahora de este tipo de predicados especiales y cómo deberían utilizarse.<br>
Todos ellos se denotan de la manera <strong>&quot;predicado/número de argumentos&quot;</strong>. <br>
Vamos a ver los más importantes y <strong>por qué</strong> deberías acordarte de ellos como del cumpleaños de tu gato.<br></p>
<h2 id="metapredicados">Metapredicados</h2>
<h3 id="call-1-el-buen-procastineo-">call/1 - El buen procastineo.</h3>
<p>El predicado <strong>call/1</strong> nos permite, sencillamente, llamar a un predicado en concreto. Si bien puede parecer redundante, nos permite poder gestionar las llamadas a predicados de manera dinámica. Es útil cuando queremos ejecutar un predicado que desconocemos en tiempo de compilación pero se determina en tiempo de compilación.<br>
Vamos, que si no lo sabemos <strong>ahora</strong> no es un problema, ya se encarga el motor de inferencia de hacer nuestro trabajo:<br></p>
<pre><code>choose_operation(Op) :-
( Op = <span class="hljs-keyword">add </span>-&gt; call(<span class="hljs-keyword">add, </span><span class="hljs-number">2</span>, <span class="hljs-number">3</span>)<span class="hljs-comment">;</span>
  Op = <span class="hljs-keyword">multiply </span>-&gt; call(<span class="hljs-keyword">multiply, </span><span class="hljs-number">2</span>, <span class="hljs-number">3</span>) ).
</code></pre><p>En este ejemplo, la operación a realizar dependerá de la operación(El valor con el que se haya ligado la variable &quot;Op&quot;).<br></p>
<p>También resulta útil desde un punto de vista de buenas prácticas, permitiendo escribir código reutilizable para distintos predicados, resultando también una programación de más alto nivel, más flexible y comprensible desde nuestro punto de vista: <br></p>
<pre><code>apply_to_all(Predicate, List) :-
    maplist(<span class="hljs-keyword">call</span>(Predicate), List).
</code></pre><p>Este predicado aplicará el predicado del primer argumento a la lista del segundo argumento. ¿Cómo? <strong>Buena pregunta</strong>, vamos con el <strong>maplist</strong>.</p>
<h3 id="maplist-2-5-trabaja-con-cabeza-no-a-cabezazos-">maplist/2-5 - Trabaja con cabeza, no a cabezazos.</h3>
<p>Este predicado aplica un predicado a todos los elementos de entre una y cuatro listas, dependiendo de nuestras necesidades y la aridad de la cabecera.<br>
Para un predicado y una lista, la aridad de maplist será dos(maplist/2), para dos listas la aridad será tres, tres listas aridad cuatro, cuatro listas aridad cinco...<br> </p>
<pre><code><span class="hljs-function"><span class="hljs-title">es_par</span><span class="hljs-params">(X)</span></span> :- <span class="hljs-number">0</span> is X mod <span class="hljs-number">2</span>.

<span class="hljs-function"><span class="hljs-title">todos_pares</span><span class="hljs-params">(List)</span></span> :-
    maplist(es_par, List).
</code></pre><p>&quot;todos_pares&quot; aplica el predicado &quot;es_par&quot; a todos los elementos de una lista, sin tener que recorrer la lista manualmente dividiéndola entre cabeza y cola ([H|T]). Esto es algo que todavía no has visto y no entenderás pero confía en mi, es <strong>Bastante</strong> más cómodo.<br></p>
<h3 id="include-3-y-exclude-3-filtro-filtro-que-es-lo-que-les-j-https-youtu-be-zttmiaf8pdi-si-osipxtpcj2jr3iui-t-8-">include/3 y exclude/3 - <a href="https://youtu.be/ZttMiaF8PDI?si=OSiPxtpCj2Jr3IUI&amp;t=8">Filtro, filtro, que es lo que les j...</a></h3>
<p>Estos dos predicados nos permiten filtrar listas basándonos en una condición, un predicado que recibe como <strong>primer</strong> parámetro. <strong>Include</strong> nos generará una lista con todos los elementos que cumplan la condición especificada, mientras que <strong>exclude</strong> lo hará con aquellos elementos que <strong>no</strong> la cumplan.<br></p>
<pre><code><span class="hljs-keyword">include</span>(Predicate, <span class="hljs-keyword">List</span>, FilteredList)./exclude(Predicate, <span class="hljs-keyword">List</span>, FilteredList).
</code></pre><p>El primer argumento será el predicado que determina la condición, el segundo la lista a la que queremos aplicar la condición y el tercer argumento será una variable que unificará con todos los elementos de la lista que cumplan la condición.</p>
<h3 id="findall-3-bagof-3-y-setof-3-soy-prologramador-deme-deme-deme-https-youtu-be-fxbvanbss-y-si-8mnhjdkeqzn4dkax-t-6-">findall/3, bagof/3 y setof/3 - <a href="https://youtu.be/FXbVANBSs-Y?si=8MNhjdkEqzn4dkaX&amp;t=6">Soy prologramador, deme deme deme.</a></h3>
<p>Ojo aquí. Esto es ya plutonio enriquecido. Estos tres predicados son literalmente la santa Trinidad de la metaprogramación en Prolog, tan pronto te curan la ceguera como te tiran cuatro meteoritos en tu casa, te inundan el váter y te convierten en una estatua de sal ahumada.<br>
El primer predicado, <strong>findall</strong> tiene tres argumentos, el elemento que queremos recoger, el predicado que ha de cumplir y la lista donde se almacenará el resultado:<br></p>
<pre><code><span class="hljs-function"><span class="hljs-title">findall</span><span class="hljs-params">(Template, Goal, List)</span></span>.
</code></pre><p><strong>Template</strong> es el elemento(Un <strong>término</strong>, y recuerda que un <strong>término</strong> puede ser CASI cualquier cosa) que queremos que se cumpla en <strong>Goal</strong> y se recoja en <strong>List</strong>.<br>
Por ejemplo:<br></p>
<pre><code><span class="hljs-function"><span class="hljs-title">menor_que_cinco</span><span class="hljs-params">(X)</span></span> :- X &lt; <span class="hljs-number">5</span>.

<span class="hljs-function"><span class="hljs-title">recoge_numeros</span><span class="hljs-params">(List)</span></span> :-
<span class="hljs-function"><span class="hljs-title">findall</span><span class="hljs-params">(X, menor_que_cinco(X)</span></span>, List).
</code></pre><p>Queremos los elementos X que sean verdaderos en &quot;menor_que_cinco(X)&quot;. &quot;findall&quot; nos devolverá todas las soluciones que cumplan esto <strong>sin</strong> ordenar, <strong>sin filtrar</strong> duplicados ni hacer nada más. Para la lista [3,1,2,5,1,7,8], findall devolverá [3,1,2,1], mientras que para la lista [6,8,6,5,8] devolverá [].<br></p>
<p>El segundo predicado, <strong>bagof</strong>, es <strong>exactamente</strong> igual que findall, <strong>PERO</strong> donde <strong>findall</strong> devolvería una lista vacía si no encuentra nada, <strong>bagof</strong> fallará si no encuentra al menos una solución. Al igual que findall, <strong>no</strong> ordena la solución y <strong>no</strong> elimina duplicados.<br></p>
<pre><code><span class="hljs-function"><span class="hljs-title">menor_que_cinco</span><span class="hljs-params">(X)</span></span> :- X &lt; <span class="hljs-number">5</span>.

<span class="hljs-function"><span class="hljs-title">recoge_numeros</span><span class="hljs-params">(List)</span></span> :-
<span class="hljs-function"><span class="hljs-title">bagof</span><span class="hljs-params">(X, menor_que_cinco(X)</span></span>, List).
</code></pre><p>Para la lista [3,1,2,5,1,7,8], findall devolverá [3,1,2,1], mientras que para la lista [6,8,6,5,8] devolverá <strong>false</strong>.<br></p>
<p>El tercer predicado, <strong>setof</strong>, tiene la misma sintaxis que sus dos hermanos más brutos, pero como lo criaron en una casa estable en lugar de entre una manada de lobos y hienas es un poco más refinado y sensato. En lugar de devolverte la solución de cualquier manera a la que te j*d€, <strong>setof</strong> ordenará primorosamente la solución para facilitarte el trabajo y <strong>además</strong> se tomará la molestia de elaborar una lista de soluciones óptima, <strong>eliminando</strong> las soluciones duplicadas y, por tanto, redundantes. Todo un fenómeno.<br></p>
<pre><code><span class="hljs-function"><span class="hljs-title">menor_que_cinco</span><span class="hljs-params">(X)</span></span> :- X &lt; <span class="hljs-number">5</span>.

<span class="hljs-function"><span class="hljs-title">recoge_numeros</span><span class="hljs-params">(List)</span></span> :-
<span class="hljs-function"><span class="hljs-title">setof</span><span class="hljs-params">(X, menor_que_cinco(X)</span></span>, List).
</code></pre><p>Para la lista [3,1,2,5,1,7,8], findall devolverá [1,2,3], mientras que para la lista [6,8,6,5,8] devolverá <strong>false</strong>.<br></p>
<h3 id="forall-2-el-absolutismo-vuelve-a-estar-de-moda-">forall/ 2 - El absolutismo vuelve a estar de moda.</h3>
<p>Este es un predicado bastante sencillo, que marida muy bien con otros metapredicados, una copa de tinto y un poco de arsénico.<br>
<strong>forall</strong> podría verse como una especialización de include/3. De manera similar a este, verificará que, para un <strong>objetivo</strong> determinado, todas las soluciones cumplan una condición determinada, especificada en forma de predicado. Si se cumple dicha condición <strong>para toda solución</strong>, devolverá <strong>true</strong>. Si no, fallará miserablemente:<br></p>
<pre><code><span class="hljs-function"><span class="hljs-title">forall</span><span class="hljs-params">(Goal, Condition)</span></span>.
</code></pre><p>En donde <strong>Goal</strong> es el término que genera soluciones, y <strong>Condition</strong> la condición a cumplir.<br></p>
<pre><code><span class="hljs-function"><span class="hljs-title">positivo</span><span class="hljs-params">(X)</span></span> :- X &gt; <span class="hljs-number">0</span>.
<span class="hljs-function"><span class="hljs-title">todos_positivos</span><span class="hljs-params">(List)</span></span> :-
    forall(member(X, List), positive(X)).
</code></pre><p>En este ejemplo, si todos los elementos de la lista son positivos, devuelve <strong>true</strong>.</p>
<h2 id="predicados-de-control-cuidado-">Predicados de control. Cuidado.</h2>
<h3 id="once-1-una-y-no-m-s-https-youtu-be-untiftqe1vq-si-af7poc1g-d0isozr-t-48-">once/1 - <a href="https://youtu.be/UntiFtQe1vQ?si=aF7PoC1G-d0isoZR&amp;t=48">Una y no más.</a></h3>
<p>Este predicado asegura que el predicado que tiene como argumento encuentra, como mucho, <strong>una</strong> solución. Esto es útil para mejorar la eficiencia de un programa en puntos en los que solo necesitamos una solución y no continuar el backtracking buscando más.<br></p>
<h3 id="-1-inocente-hasta-que-se-demuestre-lo-contrario-">\+/1  - Inocente hasta que se demuestre lo contrario.</h3>
<p>Habíamos hablado del <strong>&quot;No se puede demostrar&quot;</strong> hace cosa de <a href="../Basico/Operadores.ipynb">cuatro capítulos</a>. Como refresco de la lección anterior, recordaremos que este <strong>NO</strong> es un predicado de negación, sino que simplemente asume que algo es falso en tanto no se demuestre lo contrario. La principal ventaja es que nos permite realizar un control <strong>dinámico</strong>, pudiendo responder este predicado a situaciones en las que aparezcan nuevas reglas sin intervención nuestra.<br></p>
<h3 id="fail-0-el-caballo-de-atila-">fail/0 - El caballo de Atila.</h3>
<p>Este predicado provoca que el predicado &quot;crece_hierba(X).&quot; siempre devuelva un fallo allá por dónde pasa.<br>
Sencillamente, provoca un fallo automático en el momento en que es llamado; es el equivalente a una contradicción lógica (es falso por el mero hecho de existir).<br></p>
<h3 id="true-0-pa-ti-la-perra-gorda-">true/0 - Pa ti la perra gorda.</h3>
<p>Al contrario que <strong>fail</strong>, <strong>true</strong> siempre devuelve <strong>true</strong>.<br> 
<strong>SORPRESA</strong>.<BR>
<strong>PLOT TWIST.</strong><BR>
<strong>EMOCIÓN</strong>.<br>
Es el equivalente a una tautología lógica(Es verdadero por el mero hecho de existir).<br></p>
<h3 id="-corte">! - Corte</h3>
<p>Este predicado previene el backtracking. En el momento en que el motor de inferencia encuentra un corte, descarta todo lo anterior a ese momento y solo podrá volver hasta el punto del corte.<br>
Si bien el predicado siempre es el mismo, se suele dividir en dos en función del uso que se le de, corte <strong>rojo</strong> y corte <strong>verde</strong>.<br>
El corte <strong>rojo</strong> altera la lógica del programa, y es un recurso al que a veces se recurre para quitarse de encima bucles infinitos o problemas varios surgidos de una programación inadecuada. Este tipo de corte está bastante desaconsejado, ya que genera una grandísima cantidad de efectos secundarios y nos limita enormemente la escalabilidad de nuestras aplicaciones a causa de esto.<br>
El corte <strong>verde</strong> no altera la lógica del programa, utilizándose para mejorar tanto el uso de memoria del programa como de la eficiencia computacional del mismo. La manera más adecuada de implementarlo consiste en hacer todo el desarrollo, comprobar el correcto funcionamiento de todo y luego determinar en qué puntos un corte podría incrementar la eficiencia.<br>
Honestamente, me gustaría poder explicarte en profundidad lo que es un corte, pero probablemente saldría un tutorial con la misma longitud que este y las posibilidades de que salgas de aquí con un derrame cerebral aumentan bastante, así que mi recomendación fundamentalmente va a ser que, por el momento, te abstengas totalmente del corte.<br></p>
<h2 id="predicados-sobre-la-base-de-hechos">Predicados sobre la base de hechos</h2>
<h3 id="-dynamic">/dynamic</h3>
<p>/dynamic no es tanto un predicado como una propiedad. Podríamos verlo como un <strong>import</strong> de Java o el <strong>include</strong> de C. Simplemente habilita la manipulación dinámica(En tiempo de ejecución) de nuestra base de hechos.<br></p>
<h3 id="assert-1-asserta-1-y-assertz-1-solo-una-cosa-m-s-https-youtu-be-biw9bbwjtqu-si-mbeyqwg0hf3bhbq_-t-20-">assert/1, asserta/1 y assertz/1 - <a href="https://youtu.be/biW9BbWJtQU?si=MbeyqWG0hf3BHbq_&amp;t=20">Solo una cosa más.</a></h3>
<p>Estos tres predicados son los encargados de añadir <strong>nuevos hechos y reglas</strong> de manera dinámica(En tiempo de ejecución) a nuestra base de conocimiento.<br>
El primero, <strong>assert/1</strong>, añade la nueva cláusula al <strong>final</strong> de la base de hechos. Tiene exactamente el mismo comportamiento que <strong>assertz/1</strong> pero está <strong>DEPRECIADO</strong> y no se aconseja seguir utilizándolo, dado que no se tiene control sobre la situación en donde se inserta la cláusula.<br>
El segundo, <strong>asserta/1</strong>, añade la nueva cláusula al <strong>principio</strong> de la base de hechos.<br>
El tercero, <strong>assertz/1</strong>, añade, como ya hemos dicho, la nueva cláusula al <strong>final</strong> de la base de hechos, y es la sintaxis que se recomienda frente a <strong>assert/1</strong>.</p>
<h3 id="retract-1-y-retractall-1-estoy-cansado-jefe-https-youtu-be-_9jc1kxyya0-si-h4e8kn6jt8di-8nt-t-83-">retract/1 y retractall/1 - <a href="https://youtu.be/_9jC1kXyYA0?si=h4E8kN6jt8di-8nT&amp;t=83">Estoy cansado, jefe</a></h3>
<p>Estos dos predicados son el opuesto a la familia de <strong>assert</strong>, pero fundamentalmente tienen la misma labor aunque al revés.<br>
<strong>&quot;retract/1&quot;</strong> eliminará de forma dinámica(En tiempo de ejecución) una ocurrencia de la cláusula especificada de la base de hechos, mientras que <strong>retractall/1</strong> hará el equivalente a echarle gasolina al césped, prenderle fuego, coger las cenizas, mezclarlas con napalm, prenderles fuego otra vez, tirarlas en un campo, coger un avión, lanzar una bomba nuclear encima y despedirse saludando como la reina Victoria.<br>
Vamos, que revienta todas las cláusulas que puedan unificar con lo especificado. Una cosa a tener en cuenta es que el siguiente código:<br></p>
<pre><code><span class="hljs-function"><span class="hljs-title">retractall</span><span class="hljs-params">(predicado_importante(_)</span></span>).
</code></pre><p>Es el equivalente a hacerle un drop a la base de datos de tu puesto de trabajo y te la puede liar <strong>mal</strong>.</p>
