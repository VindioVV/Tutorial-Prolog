<h1 id="operadores-de-comparaci-n-y-operadores-de-unificaci-n-">Operadores de comparación y operadores de unificación.</h1>
<p>Le hemos perdido ya un poco el miedo a Prolog. ¿Verdad?<br>
MAL.<br>
!Vamos a recuperarlo! ¿Qué mejor para ello que profundizar un poco más en la sintaxis y hacernos la picha un lío con los operadores que se parecen a los que ya conocemos?<br>
<strong>Al lío</strong><br></p>
<p>Quizás te hayas dado cuenta de que, hasta ahora, no hemos unificado nada nosotros, sino que el motor de inferencia lo ha hecho en nuestro lugar. En Prolog, el operador de unificación es el siguiente: (Parada dramática)<br></p>
<h1 id="-">=</h1>
<p>Efectivamente, la misma cosa que utilizamos para asignar valores en el resto de lenguajes. Pero aquí <strong>no</strong> estamos asignando cosas. Estamos haciendo <strong>comparaciones</strong>, y este operador devolverá verdadero si puede unificar dos términos.<br>
Prolog es absurdamente literal. Como ya comentamos en capítulos anteriores, los átomos se interpretan <strong>tal cual</strong>. Esto da lugar a situaciones absurdas como la siguiente:<br>
Es falso que 3+2 sea igual que cinco.<br>
El operador de unificación lee las cosas tal cual están. Podrías plantarle delante un plato de huevos con patatas y un cesto con patatas fritas y dos huevos fritos en un platito y te diría que no son el mismo plato.<br>
¿Cómo afrontamos esto?<br>
Con el operador <strong>is</strong>.<br></p>
<h3 id="is">is</h3>
<p>Este operador <strong>sí</strong> resuelve las operaciones aritméticas que se le planteen antes de intentar unificar. Pero vuelvo a repetir:<br>
<strong>NO</strong><br>
<strong>HACE</strong><br>
<strong>ASIGNACIONES</strong><br>
<strong>EN PROLOG NO TENEMOS ASIGNACIONES</strong><br>
<strong>ES EL MOTOR DE INFERENCIA EL QUE DETERMINA SI UNA VARIABLE LIGA CON UN TÉRMINO</strong><BR>
Una variable puede unificarse con un valor concreto. X = 5 devolverá <strong>true</strong>, pero X no tiene un valor &quot;asignado&quot; como tal que se pueda reemplazar a voluntad. Una vez que una variable liga con otro término, esa ligadura es inquebrantable.<br>
Ahora podemos probar diferentes operaciones y comprobar que, efectivamente, Prolog es capaz de determinar si 3+2 es lo mismo que 5.<br></p>

```python
import ipywidgets as widgets
from IPython.display import display, HTML, Image
from swiplserver import PrologMQI

output_box_prolog_ejemplo_1 = widgets.Output()
input_query_prolog_ejemplo_1 = widgets.Text(
    placeholder='¡Dale chicha, majo!',
    description='Consulta:',
    disabled=False,
    continuous_update=False
)

error_count_prolog_ejemplo_1 = 0

def show_error_message_ejemplo_1(user_input_prolog_ejemplo_1):
    global error_count_prolog_ejemplo_1
    with output_box_prolog_ejemplo_1:
        output_box_prolog_ejemplo_1.clear_output()
        if error_count_prolog_ejemplo_1 < 2:
            message_ejemplo_1 = "Revisa la sintaxis, por favor."
            image_path_prolog_ejemplo_1 = './imagenes/error/error_image_1.png'
        elif error_count_prolog_ejemplo_1 < 4:
            message_ejemplo_1 = "Mira bien, ¿se te ha olvidado el punto al final? ¿El nombre del predicado es correcto?"
            image_path_prolog_ejemplo_1 = './imagenes/error/error_image_2.png'
        elif error_count_prolog_ejemplo_1 < 6:
            message_ejemplo_1 = "Estás haciendo al búho llorar."
            image_path_prolog_ejemplo_1 = './imagenes/error/error_image_3.png'
        elif error_count_prolog_ejemplo_1 < 7:
            message_ejemplo_1 = "¿Te diviertes?"
            image_path_prolog_ejemplo_1 = './imagenes/error/error_image_4.png'
        else:
            message_ejemplo_1 = "..."
            image_path_prolog_ejemplo_1 = './imagenes/error/error_image_5.png'
            img_ejemplo_1 = Image.open(image_path_prolog_ejemplo_1)
            display(img_ejemplo_1)
        
        display(HTML(f'''
            <div style="text-align: center;">
                <p>{message_ejemplo_1}</p>
                <img src="{image_path_prolog_ejemplo_1}" style="width: 200px; height: auto;">
            </div>
        '''))
        error_count_prolog_ejemplo_1 += 1

def on_key_press_prolog_ejemplo_1(event):
    global error_count_prolog_ejemplo_1
    user_input_prolog_ejemplo_1 = input_query_prolog_ejemplo_1.value.strip()
    
    with PrologMQI() as mqi:
        with mqi.create_thread() as prolog_thread_ejemplo_1:
            prolog_thread_ejemplo_1.query("set_prolog_flag(encoding,utf8).")
            try:
                query_prolog_ejemplo_1 = user_input_prolog_ejemplo_1
                prolog_thread_ejemplo_1.query_async(query_prolog_ejemplo_1, find_all=True)
                result_prolog_ejemplo_1 = prolog_thread_ejemplo_1.query_async_result()
                
                with output_box_prolog_ejemplo_1:
                    output_box_prolog_ejemplo_1.clear_output()
                    if not result_prolog_ejemplo_1:
                        show_error_message_ejemplo_1(user_input_prolog_ejemplo_1)
                    else:
                        error_count_prolog_ejemplo_1 = 0  
                        display(HTML(f'<div style="text-align: center;">"Resultado: {result_prolog_ejemplo_1}. Vamos, que es verdad."</div>'))

            
            except Exception as e:
                with output_box_prolog_ejemplo_1:
                    output_box_prolog_ejemplo_1.clear_output()
                    show_error_message_ejemplo_1(user_input_prolog_ejemplo_1)

def on_button_clicked_prolog_ejemplo_1(_):
    input_container_prolog_ejemplo_1.children = [input_query_prolog_ejemplo_1]

input_query_prolog_ejemplo_1.observe(on_key_press_prolog_ejemplo_1, names='value')

input_container_prolog_ejemplo_1 = widgets.VBox([input_query_prolog_ejemplo_1, output_box_prolog_ejemplo_1], layout=widgets.Layout(align_items='center'))
display(input_container_prolog_ejemplo_1)
```

<h1 id="-qu-m-s-operadores-tenemos-">¿Qué más operadores tenemos?</h1>
<p>Por una parte tenemos los operadores aritméticos(+,-,*,/,//(División entera),mod,^) y comparadores(&lt;,=&lt;,&gt;,&gt;=) &quot;clásicos&quot;, pero también hay que explicar algunos operadores nuevos. Ponte el cinturón, que se vienen curvas.<br></p>
<h3 id="-">==</h3>
<p>Este es el operador de igualdad sintáctica. Hay que tener cuidado con él y no confundirlo con el operador de unificación. Podemos verlo en el siguiente ejemplo:<br></p>
<pre><code><span class="hljs-attr">X</span> == <span class="hljs-number">5</span> <span class="hljs-literal">false</span>.
<span class="hljs-attr">X</span> = <span class="hljs-number">5</span> <span class="hljs-literal">true</span>.
<span class="hljs-attr">X</span> == <span class="hljs-number">5</span> <span class="hljs-literal">true</span>.
</code></pre><p>Este operador puede hacer una comparación entre términos <strong>sin</strong> unificar. Si la variable X no ha unificado con nada anteriormente, la comparación fallará.<br></p>
<h3 id="-">\==</h3>
<p>Operador de desigualdad sintáctica.<br></p>
<h3 id="-">\=</h3>
<p>Este operador se comporta como la negación de la unificación. Fallará y devolverá false si dos elementos logran unificarse.<br></p>
<h3 id="-">,</h3>
<p>Operador de conjunción. Es un Y/AND lógico. Quizás lo hayas deducido en ejemplos anteriores, pero por si acaso vamos a recalcarlo.<br></p>
<h3 id="-">;</h3>
<p>Operación de disyunción. Es un O/OR lógico. Como comentamos anteriormente, este operador debería entenderse como un recurso ocasional y no como un elemento de programación habitual. Mantengamos las buenas prácticas y usemos cláusulas de Horn siempre que sea posible.<br></p>
<h3 id="-">\+</h3>
<p>Honestamente, este no es un operador como tal, si no un <strong>predicado de control</strong>.<br>
Si, lo de que no tenemos ningún tipo de control sobre el flujo de ejecución era mentira. Pero de momento tú haz como que no has oído nada.<br>
Lo podemos llamar también &quot;No se puede demostrar&quot;. Evalúa si un objetivo no puede ser probado verdadero, y de ser así devuelve true. Sin embargo, hay que tener en cuenta que <strong>NO</strong> es un operador de negación. En lógica, la negación implica demostrar que algo es falso. El &quot;No se puede demostrar&quot; simplemente comprueba que no se puede probar que un predicado sea verdadero y por tanto asume que es falso. En cierto modo, lo podemos entender como la <strong>presunción de inocencia</strong>.<br>
Podríamos acusar a una persona de roncar, pero de no ser consciente de ello porque no tiene un volumen tan absurdo como para despertar a alguien y por ello no es consciente de sus berreas de ciervo en celo. Si quisiéramos negar que dicha persona no ronca (negación lógica) tendríamos que presentarnos con una grabadora en la habitación de esta persona y comprobar in situ si hay o no hay ronquidos. Por otra parte, haciendo uso del &quot;No se puede demostrar&quot;, al no haber testigos despertados por motosierras nocturnas podemos afirmar que, dado que no se ha probado cierto que dicha persona ronque, esta persona no ronca y es inocente de las perversas acusaciones que caen sobre su cabeza.<br></p>

```python
import ipywidgets as widgets
from PIL import Image
import io

error_images_ejemplo_1 = [
    '../Basico/imagenes/error/error_image_1.png',
    '../Basico/imagenes/error/error_image_2.png',
    '../Basico/imagenes/error/error_image_3.png',
    '../Basico/imagenes/error/error_image_4.png',
    '../Basico/imagenes/error/error_image_5.png'
]

def create_button(description, color, font_size, font_weight, width='auto'):
    button = widgets.Button(
        description=description,
        layout=widgets.Layout(width=width, padding='2px', margin='5px'),
        style={'button_color': color, 'font_weight': font_weight, 'font_size': font_size, 'text_align': 'center'}
    )
    return button


def display_image(image_path, format='png', width='300px', height='200px'):
    with open(image_path, "rb") as f:
        img = Image.open(f)
        img_byte_array = io.BytesIO()
        img.save(img_byte_array, format=format)
        img_data = img_byte_array.getvalue()
    
    return widgets.Image(value=img_data, format=format, layout=widgets.Layout(width=width, height=height))


class BaseQuizApp:
    def __init__(self, quiz_data):
        self.quiz_data = quiz_data
        self.current_question_index = 0
        self.error_count = 0
        self.output_box = widgets.Output()
        self.create_quiz()
        self.display()

    def create_quiz(self):
        pass

    def display(self):
        with self.output_box:
            self.output_box.clear_output()
            display(self.question_label)
            display(self.image_box)  
            display(self.options_box)

class QuizApp_ejemplo_1(BaseQuizApp):
    def __init__(self, quiz_data):
        super().__init__(quiz_data)

    def create_quiz(self):
        self.output_box.clear_output()
        self.option_buttons = []

        if self.current_question_index < len(self.quiz_data):
            quiz = self.quiz_data[self.current_question_index]

            self.question_label = widgets.HTML(
                value=f"<h2 style='color: #0056b3; text-align: center;'>{quiz['question']}</h2>"
            )

            self.option_buttons = [
                create_button(option, '#e8ebe9', '14px', 'bold') for option in quiz['options']
            ]

            for button in self.option_buttons:
                button.on_click(lambda b, opt=button.description: self.on_button_clicked(opt, quiz['correct_option']))

            with open('../Basico/imagenes/base/base.png', 'rb') as f:
                self.image_widget = widgets.Image(
                    value=f.read(),
                    format='png', 
                    layout=widgets.Layout(max_width='700px', max_height='700px', object_fit='contain')
                )
            self.image_box = widgets.HBox([self.image_widget], layout=widgets.Layout(justify_content='center'))

            self.options_box = widgets.VBox(self.option_buttons, layout=widgets.Layout(align_items='center'))

        else:
            self.question_label = widgets.HTML(
                value="<h2 style='color: #28a745; text-align: center;'>¡Bien hecho!</h2>"
            )
            self.options_box = widgets.VBox([])
            self.image_box = widgets.VBox([])

    def on_button_clicked(self, option, correct_option):
        if option == correct_option:
            self.show_message("Correcto", "¡Buena!")
            self.next_question()
        else:
            self.error_count += 1
            message = self.get_error_message()
            self.show_error_message(message)

    def get_error_message(self):
        if self.error_count < 2:
            return "Whops, esa no era la respuesta correcta. Prueba otra vez."
        elif self.error_count < 4:
            return "¿Estás seguro de que has atendido a la explicación?"
        elif self.error_count < 6:
            return "Estás haciendo al búho llorar."
        elif self.error_count < 7:
            return "¿Te diviertes?"
        else:
            return "..."

    def show_message(self, title, message):
        with self.output_box:
            self.output_box.clear_output()
            display(widgets.HTML(f"<h3 style='text-align: center;'>{title}: {message}</h3>"))

    def show_error_message(self, message):
        with self.output_box:
            self.output_box.clear_output()
    
            error_img_path = self.get_error_image_path()
    
            with open(error_img_path, "rb") as file:
                error_img = widgets.Image(
                    value=file.read(),
                    format='png',  
                    layout=widgets.Layout(max_width='500px', max_height='400px', object_fit='contain')
                )
    
            error_button = create_button("Venga voy", '#f44336', '14px', 'bold')
            error_button.on_click(self.retry_current_question)
    
            display(widgets.VBox([
                widgets.HBox([error_img], layout=widgets.Layout(justify_content='center')),
                widgets.HTML(f"<h4 style='text-align: center;'>{message}</h4>"),
                error_button
            ], layout=widgets.Layout(align_items='center')))

    def retry_current_question(self, b):
        self.create_quiz()
        self.display()

    def get_error_image_path(self):
        index = min(self.error_count - 1, len(error_images_ejemplo_1) - 1)
        return error_images_ejemplo_1[index]

    def next_question(self):
        self.current_question_index += 1
        self.create_quiz()
        self.display()

def run_quiz(quiz_data):
    quiz_app = QuizApp_ejemplo_1(quiz_data)
    display(quiz_app.output_box)
```
